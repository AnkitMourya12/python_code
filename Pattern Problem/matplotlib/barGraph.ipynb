```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=["python","r","ai","ml","ds"]
class_stu1=[30,10,20,25,10]
class_stu2=[40,5,20,20,10]
class_stu3=[35,5,30,15,15]
```

```python
plt.bar(classes,class_stu1,width=0.2,align="edge",color="y",edgecolor="m",linewidth=5,alpha=0.9,linestyle="--",label="class 1 student")#visible=False
```

```python
plt.barh(classes,class_stu1)
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class_stu1,width=0.6,align="edge",color="k",
        edgecolor="m",linewidth=5,alpha=0.9,linestyle="--",
        label="class 1 student")
```

```python
plt.figure(figsize=(16,9))
plt.barh(classes,class_stu1,align="edge",color="k",
        edgecolor="m",linewidth=5,alpha=0.9,linestyle="--",
        label="class 1 student")
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class_stu1,width=0.6,align="edge",color="k",
        edgecolor="m",linewidth=5,alpha=0.9,linestyle="--",
        label="class 1 student")
plt.title("Bar chart of class",fontsize=15)
plt.xlabel("classes",fontsize=15)
plt.ylabel("cno.of student",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class_stu1,width=0.2,color="b",
       
        label="class 1 student")
plt.bar(classes,class_stu2,width=0.2,color="g",label="class 2 student")

plt.title("Bar chart of class",fontsize=15)
plt.xlabel("classes",fontsize=15)
plt.ylabel("cno.of student",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))

classes_index = np.arange(len(classes))
width=0.2
plt.bar(classes_index,class_stu1,width,color="b",
       
        label="class 1 student")

plt.bar(classes_index+width,class_stu2,width,color="g",label="class 2 student")
plt.bar(classes_index+width+width,class_stu3,width,color="k",
       
        label="class 3 student")
plt.xticks(classes_index+width,classes,rotation=20)
plt.title("Bar chart of class",fontsize=15)
plt.xlabel("classes",fontsize=15)
plt.ylabel("cno.of student",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))

classes_index = np.arange(len(classes))
width=0.2
plt.barh(classes_index,class_stu1,width,color="b",
       
        label="class 1 student")

plt.barh(classes_index+width,class_stu2,width,color="g",label="class 2 student")
plt.barh(classes_index+width+width,class_stu3,width,color="k",
       
        label="class 3 student")
plt.yticks(classes_index+width,classes,rotation=20)
plt.title("Bar chart of class",fontsize=15)
plt.ylabel("classes",fontsize=15)
plt.xlabel("cno.of student",fontsize=15)
plt.legend()
plt.show()
```
